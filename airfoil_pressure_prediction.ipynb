```python
import pandas as pd
```

```python
df=pd.read_csv('airfoil_self_noise.dat',sep='\t',header=None)
df.head()
```

1. Frequency, in Hertzs.
2. Angle of attack, in degrees.
3. Chord length, in meters.
4. Free-stream velocity, in meters per second.
5. Suction side displacement thickness, in meters.

The only output is:

6. Scaled sound pressure level, in decibels.

```python
df.columns=["Frequency","Angle of attack","Chord length","Free-stream velocity","Suction side","pressure level"]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
### first do train & test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.33,random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
import pickle
```

```python
pickle.dump(regressor,open('regressor.pkl','wb'))
```

```python
regressor.score(x_train,y_train)
```

```python
# from sklearn import random forest
from sklearn.ensemble import RandomForestRegressor
```

```python
randomrorestregressor=RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
randomrorestregressor.fit(x_train,y_train)
```

```python
# get score
f"{randomrorestregressor.score(x_train,y_train)*100}%"
```

```python
# predict the test set results fro random forest
y_pred=randomrorestregressor.predict(x_test)
```

```python
# calculate test accuracy
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
# create pickle file for random forest model
pickle.dump(randomrorestregressor,open('model_random.pkl','wb'))
```
